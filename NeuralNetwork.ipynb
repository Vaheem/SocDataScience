```python
def sigmoid(a: np.ndarray):
    z = 1 / (1 + np.exp(-a))
    return z
        
        
```

```python
def forwardPassing(X: np.ndarray, W1: np.ndarray, W2: np.ndarray):
    A1 = np.dot(W1, X.T)
    Z1 = sigmoid(A1)
    Z2 = np.dot(W2, Z1)
    return A1, Z1, Z2
 
```

```python
def updateParams(W1: np.ndarray, W2: np.ndarray, dW1: np.ndarray, dW2: np.ndarray, alpha: float) -> tuple:
    W1 = W1 - alpha * dW1 
    W2 = W2 - alpha * dW2 
    return W1, W2
```

```python
def gradientDescent(X: np.ndarray, t: np.ndarray, W1: np.ndarray, W2: np.ndarray, iterations: int, alpha: float):
    for i in range(iterations):
        A1, Z1, Z2 = forwardPassing(X, W1, W2)
        dW1, dW2 = backPropagation(W1, W2, Z1, Z2, X, t)
        W1, W2 = updateParams(W1, W2, dW1, dW2, alpha)
        print("Hallo")
        print(loss(X, t, W1, W2))
    return W1, W2
```
